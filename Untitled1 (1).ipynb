```python
import pandas as pd
import random
from datetime import datetime, timedelta

# User-defined function to generate synthetic financial data
def generate_fin_data(num_records=100):
    data = []
    
    # Possible values for categorical features
    names = ["John Doe", "Jane Smith", "Alice Johnson", "Chris Brown", "Emily Davis", "Michael Wilson", 
             "Sarah Taylor", "David Anderson", "Sophia Thomas", "Daniel Harris"]
    genders = ["Male", "Female", "Other"]
    occupations = ["Engineer", "Doctor", "Teacher", "Farmer", "Business Owner", "Student", "Unemployed"]
    loan_types = ["Home Loan", "Personal Loan", "Car Loan", "Education Loan", "Business Loan"]
    sma_types = ["Type-1", "Type-2", "Type-3"]
    
    # Define tenure ranges for different loan types
    tenure_ranges = {
        "Home Loan": (120, 360),
        "Personal Loan": (60, 84),
        "Car Loan": (12, 84),
        "Education Loan": (120, 150),
        "Business Loan": (12, 60)
    }
    
    for _ in range(num_records):
        # Generate values for features
        name = random.choice(names)
        account_number = random.randint(1000000000, 9999999999)  # 10-digit account number
        age = random.randint(18, 65)
        gender = random.choice(genders)
        income = random.randint(15000, 200000)  # Monthly income in INR
        occupation = random.choice(occupations)
        loan_type = random.choice(loan_types)
        tenure = random.randint(*tenure_ranges[loan_type])  # Based on loan type
        rate_of_interest = round(random.uniform(5, 15), 2)  # Annual interest rate
        emi = round((income * rate_of_interest / 100) / 12, 2)  # Simplified EMI calculation
        emi_due_date = datetime.now() - timedelta(days=random.randint(0, 60))  # Random due date within the last 60 days
        dpd = (datetime.now() - emi_due_date).days  # Days Past Due based on current date
        collection_attempts = random.randint(0, 10)
        monthly_sma_type = random.choice(sma_types)
        
        data.append([
            name, account_number, age, gender, income, occupation, loan_type, tenure, emi, rate_of_interest, 
            emi_due_date.strftime("%d-%m-%Y"), dpd, collection_attempts, monthly_sma_type
        ])
    
    # Create a DataFrame
    columns = ["Name", "Account Number", "Age", "Gender", "Income", "Occupation", "Type of Loan", 
               "Tenure of Loan (Months)", "EMI", "Rate of Interest", "EMI Payment Due Date", 
               "Days Past Due (DPD)", "Collection Attempts", "Monthly SMA Type"]
    return pd.DataFrame(data, columns=columns)

# Generate the synthetic dataset
fin_df = generate_fin_data(num_records=100)

# Display the first few rows of the DataFrame
print(fin_df.head())
```

```python
# Define a function to assign SMA Type based on DPD
def assign_sma_type(dpd):
    if 1 <= dpd <= 30:
        return "SMA-0"
    elif 31 <= dpd <= 60:
        return "SMA-1"
    elif 61 <= dpd <= 90:
        return "SMA-2"
    else:
        return "No SMA"

# Apply the function to the DataFrame to create a new column 'SMA Type'
fin_df['SMA Type'] = fin_df['Days Past Due (DPD)'].apply(assign_sma_type)

# Display the updated DataFrame
print(fin_df[['Days Past Due (DPD)', 'SMA Type']].head())
```

```python
# Delete the column 'Monthly SMA Type'
fin_df.drop(columns=['Monthly SMA Type'], inplace=True)

# Display the updated DataFrame
print(fin_df.head())
```

```python
import pandas as pd
from datetime import datetime, timedelta
import random

# Function to simulate last 12 months' SMA status for each customer
def generate_sma_df(fin_df):
    sma_data = []
    
    for _, row in fin_df.iterrows():
        # Extract customer information
        customer_info = {
            "Name": row["Name"],
            "Account Number": row["Account Number"],
            "Age": row["Age"],
            "Gender": row["Gender"],
            "Income": row["Income"],
            "Occupation": row["Occupation"],
            "Type of Loan": row["Type of Loan"],
            "Tenure of Loan (Months)": row["Tenure of Loan (Months)"],
            "EMI": row["EMI"],
            "Rate of Interest": row["Rate of Interest"],
            "EMI Payment Due Date": row["EMI Payment Due Date"],
            "Collection Attempts": row["Collection Attempts"],
            "Monthly SMA Type": row["Monthly SMA Type"]
        }
        
        # Simulate SMA statuses for the past 12 months
        for month_offset in range(12):
            month_date = datetime.now() - timedelta(days=30 * month_offset)
            month_str = month_date.strftime("%B %Y")  # Format: Month Year
            
            # Randomly assign DPD and SMA Status for the month
            dpd = random.randint(0, 120)  # Simulate DPD for the month
            if 1 <= dpd <= 30:
                sma_status = "SMA-0"
            elif 31 <= dpd <= 60:
                sma_status = "SMA-1"
            elif 61 <= dpd <= 90:
                sma_status = "SMA-2"
            else:
                sma_status = "No SMA"
            
            # Create a record for this month
            record = {
                **customer_info,
                "Month": month_str,
                "Days Past Due (DPD)": dpd,
                "SMA Status": sma_status
            }
            sma_data.append(record)
    
    # Create the SMA DataFrame
    columns = ["Name", "Account Number", "Age", "Gender", "Income", "Occupation", "Type of Loan", 
               "Tenure of Loan (Months)", "EMI", "Rate of Interest", "EMI Payment Due Date", 
               "Days Past Due (DPD)", "Collection Attempts", "Monthly SMA Type", "Month", "SMA Status"]
    return pd.DataFrame(sma_data, columns=columns)

# Generate the SMA dataset
Sma_df = generate_sma_df(fin_df)

# Display the first few rows of the new dataset
print(Sma_df.head())
```
